```python
def imginfo(image):
    print(type(image), image.dtype, image.ndim, image.shape, image.size)
```

```python
# bgr => rgb
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # RGB 
plt.imshow(image)
# plt.show()#   
imginfo(image)
```

```python
# bgr => gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = 'gray')
imginfo(gray)
```

```python
gray[0], imginfo(gray[0])
```

```python
# gray => colormap
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_SUMMER)
plt.imshow(im_color)
imginfo(im_color)
```

```python
# gray => colormap
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
imginfo(hsv)
```

```python
#   
import numpy as np
lower_green = np.array([0, 100, 0])
upper_green = np.array([100, 200, 100])

mask_green = cv2.inRange(hsv, lower_green, upper_green)
plt.imshow(mask_green)
```

```python
rows, cols, channels = img.shape

# 
m = np.float32([[1,0,200],[0,1,100]]) # [x True, y false,  200 ], [x false, y True,  100]

dst = cv2.warpAffine(img, m, (cols,rows))

plt.imshow(dst)
```

## excute_script()
자바스크립트 함수(스크롤다운) 실행하기 execute_script("스크립트", 요소) 해당 페이지에 스크립트를 만들 때 사용합니다. 요소는 필수 파라미터는 아니고 요소가 있으면 요소에 스크립트가 실행되고 없으면 전체 페이지에 스크립트가 움직입니다.

js코드를 인자로 넣을 수 있어 브라우저 화면을 제어 할수 있게 된다. 스크롤을 하기 위해 뿐만아니라 다른 동적이벤트를 작성할 수 있다.

* 화면상 스크롤 위치 이동 : scrollTo(x,Y) ,scrollTo(x,Y+number)
* 화면 최하단으로 스크롤 이동 : scrollTo(0, document.body.scrollHeight)
* 화면을 움직이고 페이지 로드 기다리기 : time.sleep(second)
* 함수 호출시 전달인자들이 배열 형태로 저장된 객체 : arguments 
    * (arguments[0].click(), sample)
    * sample요소의 첫번째 인자에 click()을 명령

```python
#네이버 뷰는 최대 1050까지만 노출 # 스크롤 10번 자동으로 됨
for i in range(10):
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') #길이 정보를 매번 새로 가져와서 다시 인풋을 넣어줌
    time.sleep(random.randint(2, 3)) #로드 될 때까지 기다려줌
```

```python
# 블로그 타이틀, 링크, 서브텍스트 가져오기
# get_view()
views = driver.find_elements(By.CSS_SELECTOR, '.lst_view .bx') #div는 양이 많아서 더 직접적으로 잡아줌
result = []

for view in views:
    con_dict = {}
    con_dict['title'] = view.find_element(By.CSS_SELECTOR, '.title_link').text # 있는 것들을 넣어줌
    con_dict['text'] = view.find_element(By.CSS_SELECTOR, '.dsc_link').text
    con_dict['date'] = view.find_element(By.CSS_SELECTOR, '.sub').text
    result.append(con_dict)
    print(con_dict)

# 딕셔너리가 리스트 안에 들어간 형태로 출력됨
```

```python
#저장된 게시물 리스트를 데이터프레임으로 변환후 csv로 저장
df = pd.DataFrame(result)
df
```

```python
#데이터 프레임 저장
# csv는 sep 쉼표가 반드시 있어야 한글이 안깨짐
df.to_csv(f'./naver_{fname}.csv', sep=',', encoding='utf-8-sig')
```

## 요구 사항
1. 수집 개수를 입력받는다.
2. 수집 개수에 따라 페이지를 이동한다.
3. 각 페이지별 데이터를 수집한다.(타이틀 텍스트)
4. 출력

```python
#수집개수 받기
cnt = int(input('크롤링 할 건수는 몇건입니까?: '))
cnt
```

```python
#페이지 수 정하기. 몇번까지 돌려야할지. 한 페이지 당 10건임.
page_cnt = math.ceil(cnt / 10) #크롤링 할 전체 페이지 수 (올림)
print(page_cnt)
```

```python
#이중 반복문 필요: 페이지 바꾸는 반복 > 한 페이지 안에서 반복

tit_xpath = '//*[@id="search_result"]/ul/li[*]/div[1]/div[1]/a'
no = 0

for x in range(1, page_cnt+1):
    print(f'========= {x} 페이지 작업 =========')
    mylist = driver.find_elements(By.XPATH, tit_xpath) #위의 추출 코드 가져옴
    
    for item in mylist:
        no += 1        #개수를 세줌
        if no > cnt:   #요청한 개수보다 큰지 체크
            break
        print(no, item.text.replace('\n', ''))
    
    if no <= cnt:     #수집 개수가 적으면 +1해줘
        a = f'/html/body/div[3]/div/div[1]/div[15]/a[{x+1}]'
        driver.find_element(By.XPATH, a).click()
        time.sleep(2)

print('========= 작업 완료 =========')
driver.close()
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument('--window-size=974,1047')
options.add_argument('--window-position=953,0')
options.add_experimental_option("detach", True)
```

```python
import os
f_dir = 'stock'
os.makedirs(os.getcwd()+'\\' + f_dir)

URL = "http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0301"
driver = webdriver.Chrome(options=options)
driver.get(URL)
time.sleep(2)
```

##  url   

```python
news_title= []
news_url = []

for n in srclist: 
    #   (.txt)    
    title = n.text
    url = urljoin(PATH, n['href'])
    print(title, url) # 
    news_title.append(title)
    news_url.append(url)    
```

```python
news_title
```

```python
news_url
```

#   

```python
import pandas as pd
```

```python
df = pd.DataFrame({'':news_title, '':news_url})
df
```

# excel  

```python
# excel  
df.to_excel('naver.xlsx', index=False) #   
#       
```

```python
info = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
```

```python
info
```

```python
resp = requests.get(PATH, headers = info)
resp
```

```python
html_src = resp.text
html_src
```

```python
# 
soup = BeautifulSoup(html_src, 'lxml')
soup
```

```python
#  
len(soup.select('.ellipsis.rank01 a'))
```

```python
song = [item.text for item in soup.select('.ellipsis.rank01 a')]
```

```python
len(song)
```

```python
#   
# ellpsis rank02  ,   
#  2     
artist = []
for item in soup.select('.ellipsis.rank02 a') : 
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
#   >  109.   
name = [item.text for item in soup.select('.ellipsis.rank02>a')]
len(name)
```

```python
artist = []
for item in soup.select('.checkEllipsis') :
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
#  
rank_list = list(range(1, 101))
rank_list
```

```python
#     ,  
songDF = pd.DataFrame({
        "":rank_list,
        "":song,
        "":artist})
songDF
```

```python
songDF.to_excel('melon100.xlsx', index=False)
```

## import

```python
import requests
from bs4 import BeautifulSoup
```

## resp: requests 응답 정보 객체

```python
PATH = 'https://finance.naver.com/'
resp = requests.get(PATH)
```

```python
# requests 응답 정보 객체
resp
```

```python
resp?
```

## resp.text

```python
resp.text
```

```python
print(resp.text)
```

```python
src = resp.text
```

## BeautifulSoup(src, 'lxml'): 텍스트를 bs로 정리

```python
soup = BeautifulSoup(src, 'lxml') # 텍스트를 bs로 정리
print(soup)
```

## 뉴스 목록 찾기


```python
srclist = soup.select('.section_strategy a') #손자를 가져오기 위해 띄어쓰기함. 꺽쇠와 다름
srclist
```

```python
# 뉴스 목록 찾기
srclist = soup.select('.news_area a')
print(type(srclist))
```

## 리스트 개수 확인

```python
# bs태그 객체들의 리스트
print(len(srclist)) # 리스트 개수 맞는지 확인: 6개인데, 개발자가 더보기까지 넣어둠
print(srclist)
```

## 더보기 버튼 제거하기(경로지정 or 슬라이싱)

### 경로 지정해서 더보기 제외하기


```python
srclist = soup.select('.section_strategy li a')
len(srclist)
```

### 슬라이싱 사용해서 제외하기

```python
srclist = soup.select('.news_area a')[:-1]
srclist
```

## 첫번째 뉴스 데이터 추출

```python
sample = srclist[0]
```

```python
print(sample)
```

```python
# 첫번째 뉴스 중 텍스트만 추출
sample.text
```

```python
# url 추출
sample['href']
```

## 상대 URL을 절대 URL로 변환하기

```python
href = sample['href']
```

```python
# //가 중복되는 문제처럼 생각보다 합치는게 번거로울 때가 있음
PATH + href
```

```python
from urllib.parse import urljoin
```

```python
urljoin(PATH, href)
```

## 제목과 url 데이터 목록 만들기

```python
news_title= []
news_url = []

for n in srclist: 
    # 태그 객체의 메서드(.txt)를 쓰려면 태그 객체별로 빼줘야함
    title = n.text
    url = urljoin(PATH, n['href'])
    print(title, url) # 중간체크
    news_title.append(title)
    news_url.append(url)    
```

```python
news_title
```

```python
news_url
```

# 데이터 프레임 변환

```python
import pandas as pd
```

```python
df = pd.DataFrame({'제목':news_title, '주소':news_url})
df
```

# excel 파일 저장

```python
# excel 파일 저장
df.to_excel('naver.xlsx', index=False) #인덱스 번호 안 쓸게
# 경로 지정 안 하면 현재 디렉토리에 저장됨
```

# 멜론 차트, 노래제목과 가수 수집하기
https://www.melon.com/chart/index.htm

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#클라이언트 오류가 발생(발신자를 밝혀라): 내 문제구나
PATH = 'http://www.melon.com/chart/index.htm'
resp = requests.get(PATH)
resp
```

```python
html_src = resp.text
html_src
```

- ua 정보 찾기   
chrome://version/


```python
info = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
```

```python
info
```

```python
resp = requests.get(PATH, headers = info)
resp
```

```python
html_src = resp.text
html_src
```

```python
#돔으로 만들기
soup = BeautifulSoup(html_src, 'lxml')
soup
```

```python
# 곡이름 찾기
len(soup.select('.ellipsis.rank01 a'))
```

```python
song = [item.text for item in soup.select('.ellipsis.rank01 a')]
```

```python
len(song)
```

```python
# 가수 이름 찾기
# ellpsis rank02로 하면 안됨, 리스트 길이 안맞음
# 가수가 2명이상이거나 동일한 자식속성이 하나 더 있음
artist = []
for item in soup.select('.ellipsis.rank02 a') : 
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
#꺽쇠로 자식만 가져오기 > 그래도 109개. 피쳐링도 따로 잡힘
name = [item.text for item in soup.select('.ellipsis.rank02>a')]
len(name)
```

```python
artist = []
for item in soup.select('.checkEllipsis') :
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
# 인덱스 만들기
rank_list = list(range(1, 101))
rank_list
```

```python
# 사실 판다스에서 인덱스 핸들링 가능함, 다음에 알아보자
songDF = pd.DataFrame({
        "순위":rank_list,
        "노래제목":song,
        "가수명":artist})
songDF
```

```python
songDF.to_excel('melon100.xlsx', index=False)
```

# rest test site

```python
import requests
```

```python
param = {'name':'Amy', 'age':20, 'address':'Seould'}
```

```python
# 요청 정보(파라미터)가 인자에 포함되어 있다.
resp1 = requests.get('http://httpbin.org/get', params=param)
print(resp1.text)
```

```python
# 요청 정보(파라미터)가 인자에 없고, form에 숨겨져있다.(예시라서 보이는 것)
resp2 = requests.post('http://httpbin.org/post', data=param)
print(resp2.text)
```
